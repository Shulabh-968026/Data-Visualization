# Import All required library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

# Q.1 How many people smoke

```python
sns.countplot(df['smoker'])
```

# Q.2 Which Sex People Smoke More

```python
sns.countplot(x=df['smoker'],hue="sex",data=df)
```

# Q.3 Which Day People come more 

```python
sns.countplot(x=df["day"])
```

# Q.4 Which Day People give tip more

```python
sns.barplot(y="tip",x="day",data=df)
```

# Q.5 When people come more lunch or dinner 

```python
sns.countplot(x="sex",hue="time",data=df)
```

# Q.6 Which day total_bill more

```python
sns.barplot(y="total_bill",x="day",data=df)
```

# Which sex gave bill more

```python
sns.barplot(y="sex",x="total_bill",data=df)
```

# Which sex come with more people

```python
sns.barplot(x="sex",y="size",data=df)
```

# How many people come with more people

```python
sns.countplot(x=df["size"])
```

# Who give tip more smoker or non smoker

```python
sns.barplot(x="smoker",y="tip",hue="sex",data=df)
```

# Whose bill maximum with there size group

```python
sns.jointplot(x="total_bill",y="size",data=df,kind="hex")
```

# All Numeraic feature with divide by sex

```python
sns.pairplot(df,hue="sex")
```

# Who will pay more bill smoker or non smoker

```python
sns.boxplot(x="total_bill",y="smoker",data=df)
```
